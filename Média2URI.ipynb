```
a = float(input())
peso_a = 2
b = float(input())
peso_b = 3
c = float(input())
peso_c = 5
nota_a = (a*peso_a)
nota_b = (b*peso_b)
nota_c = (c*peso_c)
total = (nota_a+nota_b+nota_c)
MEDIA = (total)/(peso_a+peso_b+peso_c)
print("MEDIA = ",MEDIA)
```
