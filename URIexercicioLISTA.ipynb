```python
#lista vazia para acrescentar valores sabendo do intervalor através do RANGE
X = []
for c in range(0,10):
#acrescentando valores dentro do RANGE e atribuindo a uma variável
  valor = int(input())
#adicionando os valores de entrada dentro da lista vazia pela variável
  X.append(valor) 
#condição de que os valores de entrada iguais ou menores que zero, são substituidos por outro valor
  if valor <= 0:
     valor = 1
#presta atenção na identação do print para que não der erro na hora de imprimir.  
  print('X[{}] = {}'.format(c,valor))
#a cada {} deve ser identificada uma varivel para o print, na sequência respectiva da disposição.
```

```python
#Faça um programa que leia um vetor N[20]. 
Vetor = []
#estrutura para de loop para os 20 valores adicionando a lista vazia.
for i in range(20):
   Valor = int(input())
   Vetor.append(Valor)
#Troque a seguir, o primeiro elemento com o último, o segundo elemento com o penúltimo.
#Invertendo a ordem, deve criar outra variavel agora com uma lista invertida. 
#[::-1] isso é um slice conteúdo que não foi visto durante as aulas.
A = Vetor[::-1]
for c in range(20):
    print('Vetor[{}] = {}'.format(c,A[c]))

#código diferente do outro devido a identação, não é foi possível observar diferença a não ser essa.

```
