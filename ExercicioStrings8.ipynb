```python
#subtrair o primeiro numero caso exista no segundo numero em uma mesma string, convertando em int para reduzir os zeros caso exista.
#usando a flag para interromper while true caso ambos forem 0
while True:
  a, b = input().split()
  if a == '0' and b =='0':
    break

  b = '0' + b
#em b sera substituido o número de a em valor "apagado."
  val = int(b.replace(a,''))
  print(val)
 
  
```

```python
#somar dois valores da mesma string, apois a soma, se existir um zero no resultado, excluir ele do print.
#condição de interromper o laço os valores forem 0
while True:
    a, b = input().split()
    if(a == '0' and b == '0'):
        break
# a soma dos valores de A e B
    a = int(a) + int(b)
#convertendo a string em inteiro e substituindo o número zero do resultado.
    val = int(str(a).replace('0',''))
    print(val)
```

```python
#digitar 2 números como texto e saber se o segundo encaixa no fim do primeiro.
#casos para testes
testes = int(input())

#laço do tamanho dos casos, tendo os números em uma mesma entrada e dividindo em duas variavéis,
for i in range(testes):
  texto = str(input())
  entradas = texto.split()
  x = entradas[0]
  y = entradas[1]

#condição para saber se  Y encaixa ou não, é fatiando uma string pela outra.  
  if x[len(x)-len(y):] ==y:
    print('encaixa')
  else:
    print('nao encaixa')
  
```

```python
C = int(input())
for i in range(C):
  B, E = input().split()
  valor = ''
  for d in range (B, E+1):
    valor += str(d)
  print(valor + valor[::-1])
```

```python
c = int(input())
for i in range(c):
    saida = ""
    reverse = ""
    a, b = map(int, input().split())
    while a <= b:
        saida += str(a)
        a += 1

    reverse = saida + saida[::-1]
    print(reverse)
```

```python
C = int(input())
for i in range(C):
  B, E = input().split()
  inicio = ''
  espelho = ''
  while B <= E:
    inicio += str(B)
    B += '1'
  espelho = inicio + inicio[::-1]
  print(espelho)
```
