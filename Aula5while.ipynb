```
par = impar = 0
positivo = negativo = 0
for n in range (1,6):
    n = int(input("Digite um número: "))
    if n % 2 == 0:
        par = par + 1
    if n % 2 != 0:
        impar = impar + 1
    if n > 0:
        positivo = positivo + 1
    if n < 0:
        negativo = negativo + 1
print(par,"valor(es) par(es)")
print(impar,"valor(es) impar(es)")
print(positivo,"valor(es) positivo(s)")
print(negativo,"valor(es) negativo(s)")

#código para ler números e contar quantos valores foram inseridos conforme as condições
#par, impar, positivo, negativo
```

```
#atribuindo valores a serem inseridos
x = int(input())
y = int(input())
# condição onde atenderá os critérios de maior e menores números podendo ser x ou y
maior = x if x > y else y
menor = y if y < x else x
# condição inicial, onde o menor número já iniciará + 1.
menor+=1
soma = 0
# enquanto menor < maior, ou seja até que o numero do loop não ultrapasse o maior inserido (x ou y)
while (menor < maior):
# menor sendo impar, será dada a soma + menor.
    if(menor % 2 != 0):
        soma+=menor
    menor+=1
print(soma)
# código para ler 2 numeros e dentro do intervalo entre eles, somar os numeros impares.
```
